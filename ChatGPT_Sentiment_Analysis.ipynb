```python
from textblob import TextBlob
df = pd.read_csv('tweets.csv',encoding = 'utf8')
df.head()
text_data = df['text']

text_data
```

```python
# Check for missing values and replace them with an empty string
text_data = text_data.fillna('')
```

```python
# Initialize the VADER model
sia = SentimentIntensityAnalyzer()
```

```python
# Perform sentiment analysis on the text data
sentiment_scores = []
for text in text_data:
    score = sia.polarity_scores(text)['compound']
    sentiment_scores.append(score)
```

```python
# Add the sentiment scores as a new column in the DataFrame
df['sentiment_score'] = sentiment_scores
```

```python
text_d = df['sentiment_score']
# Check for missing values and replace them with an empty string
text_d = text_data.fillna('')
```

```python
sia = SentimentIntensityAnalyzer()

sentiments = []
for text in text_d:
    score = sia.polarity_scores(text)['compound']
    if score > 0.05:
        sentiments.append('positive')
    elif score < -0.05:
        sentiments.append('negative')
    else:
        sentiments.append('neutral')
```

```python
# Add the sentiments as a new column in the DataFrame
df['sentiment'] = sentiments
```
